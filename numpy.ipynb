```python
t = arange(0.,5.,0.2)

plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'ro', label='^2.2')
plt.plot(t, t**2.5, 'g^',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [1,3,5,7]
x3 = [1,2,3,4]
y3 = [2,4,6,8]

l = plt.plot(x1,y1,x2,y2,x3,y3)
plt.setp(l[0],color = 'r',linewidth = 2.0)
plt.setp(l[1],color = 'b',linewidth = 2.0)
plt.setp(l[2],color = 'g',linewidth = 2.0)
plt.grid()
plt.show()
```

```python
indian_temp = pd.DataFrame({
    'city' : ['Andhra','Telangana','Bangalore'],
    'temperature' : [35,40,28],
    'Humidity' : [25,45,20]
})
indian_temp
```

```python
us_temp = pd.DataFrame({
    'city' : ['chicago','newjersey','newyork'],
    'temperature' : [20,22,24],
    'Humidity' : [10,20,30]
})
us_temp
```

```python
pd.concat([indian_temp,us_temp])
```

```python
pd.concat([indian_temp,us_temp],ignore_index=True)
```

```python
pd.concat([indian_temp,us_temp],axis=1)
```

```python
temp_df = pd.DataFrame({
    'city' : ['andhra','Telangana','Bangalore','Chennai'],
    'temperature' : [20,25,30,35]
})
temp_df
```

```python
humidity_df = pd.DataFrame({
    'city' : ['andhra','Telangana','Bangalore'],
    'humidity' : [10,15,20]
})
humidity_df
```

```python
pd.merge(temp_df,humidity_df,on='city')
```

```python
pd.merge(temp_df,humidity_df,on='city',how='outer')
```

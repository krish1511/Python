```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(10,20)
print(x)
print(y)
```

```python
a = np.arange(20,30)
b = np.arange(30,40)
print(a)
print(b)
```

```python
plt.scatter(x,y,c='g')    # 2 dimensional visulaization plot

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y, 'r*--', linestyle = 'dashed', linewidth = 2, markersize = 12)
```

```python
y = x*x
```

```python
plt.plot(x,y, 'r*--', linestyle = 'dashed', linewidth = 2, markersize = 12)
plt.xlabel('x axis')
plt.ylabel('y label')
plt.title('Graph in 2D')
```

```python
# Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 4* np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form')

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,5 * np.pi , 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g')
plt.title('cosine')

plt.show()
```

```python
# Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align='center',color='y')
plt.bar(x2,y2,color='g')
plt.title('bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
# Histograms  # it can be plotted to single variable
a = np.random.randint(10,100,10)
print(a)
plt.hist(a)
plt.title('histogram')
plt.show()
```

```python
# Box plot using matplotib

data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data, vert=True,patch_artist=True)
```
